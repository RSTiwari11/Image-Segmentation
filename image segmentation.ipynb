```python
segmentation_pascal = semantic_segmentation()
segmentation_ade = semantic_segmentation()
help(segmentation_pascal)
help(segmentation_ade)
```

```python
path_image="drive/My Drive/Segmentation_dfiles/images (1).jpg"
segmentation_ade.load_ade20k_model(path_adek)
segmentation_pascal.load_pascalvoc_model(path_pascal)
```
