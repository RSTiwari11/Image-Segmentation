```python
from pixellib.instance import instance_segmentation
instance_segment = instance_segmentation()
instance_segment.load_model(path_coco)
```

```python
path_vid=path+"/com.mp4"
A=0
cap = cv2.VideoCapture(path_vid)
frame_video=[]
segmented_instanced=[]
if (cap.isOpened()== False): 
    print("Error opening video stream or file")

while(cap.isOpened()):
    ret, frame = cap.read()

    if ret == True:
        
        print(A)
        A=A+1
        
        frame=cv2.resize(frame,(600,600))
        frame_video.append(frame)
        #cv2.resizeWindow("output", 800,400)
        #cv2.namedWindow("output", cv2.WINDOW_NORMAL)
        segmask, output_inst = instance_segment.segmentFrame(frame, show_bboxes=False, output_image_name=None)
        segmented_instanced.append(output_inst)
        #cv2.imshow("output",frame)
    
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else: 
        break

cap.release()
cv2.destroyAllWindows()
```

```python
segmented_instanced =np.asarray( segmented_instanced )
print( segmented_instanced .shape)
```
